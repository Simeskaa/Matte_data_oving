# Vektorfelt i python
Vi skal se på hvordan vi kan bruke Python for å løse ulike oppgaver relatert til vektorfelt. Først ser vi på hvordan vektorfelt blir tegnet opp.

For å kunne tegne opp et vektorfelt trenger vi en følge med x- og y-verdier. Vi kan for eksempel gjøre det via metodene "linspace" og "arange" i numpy. 

```python
import numpy as np

xp = np.linspace(-3, 3, 13) # tallfølge som starter på -3 og slutter på 3. Har totalt 13 punkter
yp = np.arange(-3, 3.5, 0.5) # Her starter vi også på -3, men øker verdien på neste tall med 0.5 helt til neste verdi vil være lik eller større enn maksverdi

# vi kan skrive ut tallrekkene og se hva de inneholder
print("xverdier laget med linspace: ", xp)
print("\nyverdier laget med arange: ", yp)
```

Både "linspace" og "arange" er nyttige på sin måte. "linspace" gir mer kontroll over hvor mange punkter vi har, mens "arange" gjør det lettere å kontrollere avstanden mellom punktene. 

Lær dere bruken til minst en av disse metodene, og velg en metode som dere bruker gjennom øvingen.

Vi bruker nå numpys "meshgrid" til å lage en mengde punkter i 2d ved hjelp av x og y-verdiene vi har laget over. 

```python
# Vi trenger x og y-verdiene til punktene (vi laget også disse over)
# Vi bruker "linspace" her, men du kan godt bruke "arange" istedet.
xverdier = np.linspace(-3, 3, 13)
yverdier = np.linspace(-3, 3, 13)

# meshx og meshy tilsammen gir oss en mengde med punkter i planet
xp, yp = np.meshgrid(xverdier, yverdier)

# x og y-verdiene til punktene
print("x-verdiene til punktene:\n", xp)
print("y-verdiene til punktene:\n", yp)
```

For å lage et vektorfelt trenger vi en rekke punkter og en funksjon som beskriver vektoren i hvert punkt. 

Under lager vi et konstant vektorfelt med vektor $(u,v)=(1,0)$ i hvert punkt $(x,y)$. Vi kan deretter bruke "matplotlib.pyplot.quiver" for å tegne vektorfeltet. 

```python
import matplotlib.pyplot as plt
%matplotlib inline

u = 1
v = 0

fig = plt.figure(figsize=(5,5))
# plotter vektoren (u,v)=(1,0) i punktene (xp,yp)
plt.quiver(xp,yp,u,v)
plt.show()
```

Ved å legge inn funksjonsuttrykk for $u$ og $v$ så kan vi lage mer kompliserte vektorfelt. Under så plotter vi vektorfeltet $(Fx,Fy)=(-y^2+2, x^2-1)$. 

```python
# x^2 skrives som x**2 i python
u = -yp**2+2
v = xp**2 - 1

fig = plt.figure(figsize=(5,5))
# plotter vektoren (u(x,y), v(x,y)) i punktet (x,y)
plt.quiver(xp, yp, u, v)
plt.show()
```

Noen vektorer i vektorfeltet er for korte for å kunne se retningen som de peker i. Vi ønsker nå får vi et bedre inntrykk av retningene til vektorene i vektorfeltet. Derfor normaliserer vi vektorene. Under plotter vi et vektorfelt av enhetsvekorene $(\hat{F}x,\hat{F}y)=(-y^2+2, x^2-1)/\sqrt{(-y^2+2)^2+(x^2-1)^2}$.

```python
# ue er enhetsvektoren som peker i samme retning som u
ue = u/np.sqrt(u**2+v**2)
ve = v/np.sqrt(u**2+v**2)

fig = plt.figure(figsize=(5,5))
# plotter vektoren (u(x,y), v(x,y)) i punktet (x,y)
plt.quiver(xp, yp, ue, ve)
plt.show()
```

Vi skal nå se på hvordan vi kan plotte feltlinjer/integralkurver for feltet $(u,v)$ vi plottet over. Vi velger et vilkårlig startpunkt og integrerer numeriske med små tidssteg $\Delta t$. For hvert steg integrasjonen får vi et nytt punkt $(x+\Delta t \cdot u(x,y), y+\Delta t \cdot v(x,y)$, og vi tegner tilslutt kurven som går gjennom alle punktene. Dette gir da (en tilnærming til) integralkurven som går gjennom punktet vi startet med. Dette er også banen til en partikkel som beveger seg gjennom feltet, hvis feltet er et hastighetsfelt.  

```python
N = 110 # antall steg vi tar
deltat = 0.05 # delta t

feltlinjex = [1.8] # x-koordinat for startsposisjon 
feltlinjey = [-2.2] # y-koordinat for startposisjon

for i in range(N): # integrasjonen
    # feltlinjex[-1] og feltlinjey[-1] er det forrige punktet vi regnet ut
    xf = feltlinjex[-1]
    yf = feltlinjey[-1]
    
    # vi regner ut feltet i punktet (xf, yf)
    uf = (-yf**2 + 2)
    vf = xf**2 - 1
    
    # vi regner ut endring i posisjon ved et tidssteg (lineær tilnærming)
    deltax = uf * deltat 
    deltay = vf * deltat
    
    # vi legger det nye punktet til listen av punkt
    feltlinjex.append(deltax + xf)
    feltlinjey.append(deltay + yf)

# framvisning av vektorfelt og bevegelse til et partikkel
plt.quiver(xp, yp, u, v)
plt.plot(feltlinjex, feltlinjey)
```

Vi kan også tegne banen i en annen farge med et ekstra argument i matplotlib.pyplot.plot

```python
plt.quiver(xp, yp, u, v)
plt.plot(feltlinjex, feltlinjey, color="orange")
```

Utregningen av banen til en partikkel over ble utført ved å integrere numerisk. Python tilbyr utregning av slike integraler gjennom Scipy.integrate.odeint:

```python
from scipy.integrate import odeint

def f(punkt, t, A, B): # vi må definere en funksjon f som regner ut neste steg i integrasjonen
    x, y = punkt
    deltax = (-y**2 + A)
    deltay = x**2 - B
    return [deltax, deltay]

t = np.linspace(0, 6, 101) # definerer tidsintervallet vi ønsker å integrere over

sol = odeint(f, [1.8, -2.2], t, args=(2, 1)) # gjennomfører integrasjon. args=(2,1) betyr at A = 2 og B = 1 for funksjonen f beskrevet over

plt.plot(sol[:, 0], sol[:, 1], color="orange")
plt.quiver(xp, yp, u, v)
```

Integralkurvene kan også skisseres ved å bruke matplotlib.pyplot.streamplot. 

```python
plt.streamplot(xp, yp, u, v)
```

## Nivåkurver i python

Matplotlib inkluderer også metoder for å tegne opp høydekurver/nivåkurver. Nivåkruver vil se ut som et 2d plot av en 3d funskjon der $ z(x,y) = c $ 

For å sette opp disse nivåkurvene i python trenger vi varierende verdier i x og y retning samt en funksjon $z(x,y)$.

```python
# definerer z(x, y). z må være et 2d meshgrid
zp =  (-yp**2 + 2) + (xp**2 - 1)

# Tegner andre plot med spesifiserte nivåkurver z = 0.2, 1.0, 2, 4
plt.contour(xp,yp,zp, [0.2, 1.0, 2, 4])
```
